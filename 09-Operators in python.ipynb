# Operators in Python
#### Operators are used to perform operation on variable and values
- Arithmetic Operators  
- Relational Operators or comparison operator  
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

### Arithmetric Operators

```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

### Relational Operators

```python
# Relational Operators
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
# Logical Operators
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

print(~3)

print(4 >> 2)

print(5 << 2)
```

```python
# Assignment Operators

a = 1
a +=1
print(a)

b = 1
b -=1
print(b)

c = 1
c /=1
print(c)

d = 1
d *=1
print(d)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input('Enter a 3 digit number'))

# 345%10 -> 5
a = number%10
number = number//10

# 34%10 -> 4
b = number % 10
number = number//10

# 3 % 10 -> 3
c = number % 10

print(a + b + c)
```
