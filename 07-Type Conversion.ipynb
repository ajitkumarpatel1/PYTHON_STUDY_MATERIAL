# Type Conversion:
* Type Conversion is a process by which we change the type of a data to another types,if possible.
* Types convertion basically happen 2 types
1.   Implicit - Python automaticaly detect and convert the datatypes.
2.   Explicit - User need to change the data types manualy.

```python
# Implicit

print(5+5.6)
print(type(5),type(5.6))
```

```python
# Not possible because for the data types
# print(4 + '4')
```

```python
# Explicit
# *********************************

# str -> int
print(int("5"))

# int to str
print(str(5))

# int to float
print(float(4))
```
