### b)Accessing Substrings from a String

```python
# INDEXING
# Positive Indexing
s = 'hello world'
print(s[4])
# print(s[41])
```

```python
# Negative Indexing
# Grab the element at the index -1, which is the LAST element
s = 'hello world'
print(s[-1])
print(s[-3])
```

```python
# SLICING
# Assign string as a string
string  = 'Principal Component Analysis!'
print(string[:13])
```

```python
# Grab everything starting from index 10 till index 18
print(string[10:])   
#If you do not specify the ending index, then all elements are extracted which 
#comes after the starting index including the element at that starting index. 
#The operation knows only to stop when it has run through the entire string.
```

```python
print(string[3:5])
```

```python
#Everything
print(string[:])
```

```python
# Reverse the string
print(string[::-1])   #start, stop, stepsize
```

### Operations on Strings

- Arithmetic Operations
- Relational Operations
- Logical Operations
- Loops on Strings
- Membership Operations

```python
# Arithmetic Operation  string( + , * )


# string concatination (addation)
print('delhi' + ' ' + 'mumbai')

# string multiplication
print("*"*50)

print('delhi' + '4' + 'mumbai')
```

```python
# Relational Operator (compariosn operator) (< ,> , <=, >=, ==, !=)

# Graterthan
print('mumbai' > 'pune') # lexiographically
print('Pune' > 'pune')

# Grater than equalto >=
print('mumbai' >= 'pune')

# Smaller than equalto >=
print('mumbai' <= 'pune')
print('mumbai' <= 'pune')
```

```python
# Logical Operations

print('hello' and 'world')

print('hello' or 'world')

print(not 'hello')
```

```python
# Loops in string
for i in 'hello':
  print(i)
```

```python
for i in 'delhi':
  print('pune')
```

```python
# Membership operator
my_string = 'Albert Einstein'
print('Albert' in my_string)

print('Albert' not in my_string)
```

## String method
#### a)Capitalize/Title/Upper/Lower/Swapcase

```python
s = 'hello world'
print(s)


print("capitalized: " ,s.capitalize())        #capitalize() method convert the only 1st latter of word to capital

print("title: " ,s.title())                   #title() method convert the 1st latter of ever word  to capital

print("upper: " ,s.upper())                   #upper() method converts the string to uppercase

print("lower:",'Hello Wolrd'.lower())         #lower() method converts the string to lowercase

print("swapcase:",'HeLlO WorLD'.swapcase())   #swapcase() method convrt lower case to upeer and upper to lower
```

#### b)Count/Find/Index

```python
# COUNT
# count() method returns the count of a string in the given string. 
# Unlike lower() and upper() method, the count() method takes a string as an argument

print('my name is ajit'.count('i'))
```

```python
# FIND
# find() method returns the index of the first occurrence of a string present in a given string. 
# Similar to the count() method, the find() method takes a string as an argument

'my name is ajit'.find('x')
```

```python
#INDEX
# index() method returns the index of the first occurrence of a string present in a given string.
'my name is AJIT'.index('A')
'my name is AJIT'.index('X')  # not present - error
```
