```python
# F-STRING
first_name = 'Rahul'
last_name = 'Modi'

full_name = f'Left plus right makes {first_name}  {last_name}'  # Use {} to print the variable you want to
print(full_name)
```

```python
# FORMAT (string method)
name = 'ajit'
gender = 'male'

print('Hi my name is {1} and I am a {0}'.format(gender,name))  # positional argument
print('Hi my name is {} and I am a {}'.format(gender,name))    # defult argument
```
