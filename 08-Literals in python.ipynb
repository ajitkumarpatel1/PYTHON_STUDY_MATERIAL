# Literals in python:
 literals are devided into many types but this are some important literals:
                    
          a) Numerical literals: Binary, Octal, Decimal, Hexadecimal, Flaot
          b) String literals :
          c) Boolean literals:
          d) Special literals:

```python
a = 0b1010 #Binary Literals
b = 100 #Decimal Literal 
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5 
float_2 = 1.5e2 # 1.5 * 10^2
float_3 = 1.5e-3 # 1.5 * 10^-3

#Complex Literal 
x = 3.14j

print(a, b, c, d)
print(float_1, float_2,float_3)
print(x, x.imag, x.real)
```

```python
# binary
x = 3.14j
print(x.imag)
```

```python
# string literals
string = 'This is Python'
strings = "This is Python"
char = "C"
multiline_str = """This is a multiline string with more than one line code."""
unicode = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
```

```python
# boolian literals
a = True + 4
b = False + 10

print("a:", a)
print("b:", b)
```
