# 1.Python Data Types
* There are basically 3  types of data types are present in python


1.   Basic Data Types : - (integer, float, boolean, complex, and string etc..) 
2.   Container Data Types: - (list, Tuple, sets, Dictonary etc..)
3.   User-Defined Data Types:- (Class (oop) helps to create this)



### a)BASIC DATA TYPES:

```python
## In python precession integers number are used it means . (It means any no of integer we can use)
## Integer in python have a limit of 1e308 (1*10^308). after this no python shows this as inf

# Integer
print(8)
# 1*10^308
print(1e309)
```

```python
## Decimal/Float in python have a limit of 1e308 (1*10^308). after this no python shows this as inf
# Decimal/Float
print(8.55)
print(1.7e309)
```

```python
# Boolean
print(True)
print(False)
```

```python
# To write string we can use ('',"",""" """)
# Text/String
print('Hello World')
```

```python
# complex
print(5+6j)
```

### b)CONTAINER DATA TYPES:

```python
# List-> C-> Array
print([1,2,3,4,5])
```

```python
# Tuple
print((1,2,3,4,5))
```

```python
# Sets
print({1,2,3,4,5})
```

```python
# Dictionary
print({'name':'Nitish','gender':'Male','weight':70})
```
